```python
def sucs_grainsize(liquid_limit: float, plastic_limit: float, gravel: float, sand: float, is_organic: str):
    '''
    | Does the ASTM D324 soil clasifficacion
    | But only the fine grain soils for now
    '''
    plasticity_index = liquid_limit - plastic_limit
    retained_sieve_200 = gravel + sand
    fines = 100 - retained_sieve_200
    sucs_class = 'no class'
    sucs_group = 'no group'
    
    #FOR FINE GRAINED SOILS
    if retained_sieve_200 <= 50: #50% or more passes sieve no.200
        # FOR LL < 50
        if liquid_limit < 50:
            # For organic soils
            if is_organic == 'true':
                sucs_group = 'Group symbol: OL'
                if (plasticity_index >= 4 and plasticity_index >= (0.73*(liquid_limit - 20))):
                    if retained_sieve_200 < 30:
                        if retained_sieve_200 < 15:
                            sucs_class = 'Organic clay'
                        else:
                            if sand >= gravel:
                                sucs_class = 'Organic clay with sand'
                            else:
                                sucs_class = 'Organic clay with gravel'
                    else:
                        if sand >= gravel:
                            if gravel < 15:
                                sucs_class = 'Sandy organic clay'
                            else:
                                sucs_class = 'Sandy organic clay with gravel'
                        else:
                            if sand < 15:
                                sucs_class = 'Gravelly organic clay'
                            else:
                                sucs_class = 'Gravelly organic clay with sand'
                elif (plasticity_index < 4 or plasticity_index < (0.73*(liquid_limit - 20))):
                    if retained_sieve_200 < 30:
                        if retained_sieve_200 < 15:
                            sucs_class = 'Organic silt'
                        else:
                            if sand >= gravel:
                                sucs_class = 'Organic silt with sand'
                            else:
                                sucs_class = 'Organic silt with gravel'
                    else:
                        if sand >= gravel:
                            if gravel < 15:
                                sucs_class = 'Sandy organic silt'
                            else:
                                sucs_class = 'Sandy organic silt with gravel'
                        else:
                            if sand < 15:
                                sucs_class = 'Gravelly organic silt'
                            else:
                                sucs_class = 'Gravelly organic silt with sand'
                else:
                    print('Code stoped here AAAAA')
            # For no organic materials
            else:
                if (plasticity_index > 7 and plasticity_index >= (0.73*(liquid_limit - 20))):
                    sucs_group = 'Group symbol: CL'
                    if retained_sieve_200 < 30:
                        if retained_sieve_200 < 15:
                            sucs_class = 'Lean clay'
                        else:
                            if sand >= gravel:
                                sucs_class = 'Lean clay with sand'
                            else:
                                sucs_class = 'Lean clay with gravel'
                    else:
                        if sand >= gravel:
                            if gravel < 15:
                                sucs_class = 'Sandy lean clay'
                            else:
                                sucs_class = 'Sandy lean clay with gravel'
                        else:
                            if sand < 15:
                                sucs_class = 'Gravelly lean clay'
                            else:
                                sucs_class = 'Gravelly lean clay with sand'
                elif (4 <= plasticity_index <= 7 and plasticity_index >= (0.73*(liquid_limit - 20))):
                    sucs_group = 'Group symbol: CL-ML'
                    if retained_sieve_200 < 30:
                        if retained_sieve_200 < 15:
                            sucs_class = 'Silty clay'
                        else:
                            if sand >= gravel:
                                sucs_class = 'Silty clay with sand'
                            else:
                                sucs_class = 'Silty clay with gravel'
                    else:
                        if sand >= gravel:
                            if gravel < 15:
                                sucs_class = 'Sandy silty clay'
                            else:
                                sucs_class = 'Sandy silty clay with gravel'
                        else:
                            if sand < 15:
                                sucs_class = 'Gravelly silty clay'
                            else:
                                sucs_class = 'Gravelly silty clay with sand'
                elif (plasticity_index < 4 or plasticity_index < (0.73*(liquid_limit - 20))):
                    sucs_group = 'Group symbol: ML'
                    if retained_sieve_200 < 30:
                        if retained_sieve_200 < 15:
                            sucs_class = 'Silt'
                        else:
                            if sand >= gravel:
                                sucs_class = 'Silt with sand'
                            else:
                                sucs_class = 'Silt with gravel'
                    else:
                        if sand >= gravel:
                            if gravel < 15:
                                sucs_class = 'Sandy silt'
                            else:
                                sucs_class = 'Sandy silt with gravel'
                        else:
                            if sand < 15:
                                sucs_class = 'Gravelly silt'
                            else:
                                sucs_class = 'Gravelly silt with sand'
                else:
                    print('here is where the code breaks BBBBBB')
        # FOR LL >= 50
        elif liquid_limit >= 50:
            # For organic soils
            if is_organic == 'true':
                sucs_group = 'Group symbol: OH'
                if plasticity_index >= (0.73*(liquid_limit - 20)):
                    if retained_sieve_200 < 30:
                        if retained_sieve_200 < 15:
                            sucs_class = 'Organic clay'
                        else:
                            if sand >= gravel:
                                sucs_class = 'Organic clay with sand'
                            else:
                                sucs_class = 'Organic clay with gravel'
                    else:
                        if sand >= gravel:
                            if gravel < 15:
                                sucs_class = 'Sandy organic clay'
                            else:
                                sucs_class = 'Sandy organic clay with gravel'
                        else:
                            if sand < 15:
                                sucs_class = 'Gravelly organic clay'
                            else:
                                sucs_class = 'Gravelly organic clay with sand'
                elif plasticity_index < (0.73*(liquid_limit - 20)):
                    if retained_sieve_200 < 30:
                        if retained_sieve_200 < 15:
                            sucs_class = 'Organic silt'
                        else:
                            if sand >= gravel:
                                sucs_class = 'Organic silt with sand'
                            else:
                                sucs_class = 'Organic silt with gravel'
                    else:
                        if sand >= gravel:
                            if gravel < 15:
                                sucs_class = 'Sandy organic silt'
                            else:
                                sucs_class = 'Sandy organic silt with gravel'
                        else:
                            if sand < 15:
                                sucs_class = 'Gravelly organic silt'
                            else:
                                sucs_class = 'Gravelly organic silt with sand'
                else:
                    print('Code stoped here CCCCC')
            # For no organic materials
            else:
                if (plasticity_index >= (0.73*(liquid_limit - 20))):
                    sucs_group = 'Group symbol: CH'
                    if retained_sieve_200 < 30:
                        if retained_sieve_200 < 15:
                            sucs_class = 'Fat clay'
                        else:
                            if sand >= gravel:
                                sucs_class = 'Fat clay with sand'
                            else:
                                sucs_class = 'Fat clay with gravel'
                    else:
                        if sand >= gravel:
                            if gravel < 15:
                                sucs_class = 'Sandy fat clay'
                            else:
                                sucs_class = 'Sandy fat clay with gravel'
                        else:
                            if sand < 15:
                                sucs_class = 'Gravelly fat clay'
                            else:
                                sucs_class = 'Gravelly fat clay with sand'
                elif (plasticity_index < (0.73*(liquid_limit - 20))):
                    sucs_group = 'Group symbol: MH'
                    if retained_sieve_200 < 30:
                        if retained_sieve_200 < 15:
                            sucs_class = 'Elastic clay'
                        else:
                            if sand >= gravel:
                                sucs_class = 'Elastic clay with sand'
                            else:
                                sucs_class = 'Elastic clay with gravel'
                    else:
                        if sand >= gravel:
                            if gravel < 15:
                                sucs_class = 'Sandy elastic clay'
                            else:
                                sucs_class = 'Sandy elastic clay with gravel'
                        else:
                            if sand < 15:
                                sucs_class = 'Gravelly elastic clay'
                            else:
                                sucs_class = 'Gravelly elastic clay with sand'
                else:
                    print('here is where the code breaks DDDDDDDDDD')
        else:
            print('here is where the code breaks EEEEEEFINAL')
    
    #FOR COARSE GRAINED SOILS
    else: #More than 50% retained in sieve no.200
        if (gravel > sand):
            if fines < 5:
                print('this branch uses Cu/Cc and is not coded yet')
            elif (5 <= fines <= 12):
                print('this branch uses Cu/Cc and is not coded yet')
            else: #fines >12%
                #this basically says: if ML or MH (under A Line)
                if (plasticity_index < 4 or plasticity_index < (0.73*(liquid_limit - 20))):
                    sucs_group = 'Group symbol: GM'
                    if sand < 15:
                        sucs_class = 'Silty gravel'
                    else: #sand>=15%
                        sucs_class = 'Silty gravel with sand'
                #this basically says: if CL or CH (above A line)
                elif (plasticity_index > 7 and plasticity_index >= (0.73*(liquid_limit - 20))):
                    sucs_group = 'Group symbol: GC'
                    if sand < 15:
                        sucs_class = 'Clayey gravel'
                    else: #sand>=15%
                        sucs_class = 'Clayey gravel with sand'
                #this basically says: if CL-ML (grey box)
                elif (4 <= plasticity_index <= 7 and plasticity_index >= (0.73*(liquid_limit - 20))):
                    sucs_group = 'Group symbol: GC-GM'
                    if sand < 15:
                        sucs_class = 'Silty, clayey gravel'
                    else: #sand>=15%
                        sucs_class = 'Silty, clayey gravel with sand'
                else:
                    print('breaks here FFFF')
        else: #sand > gravel
            if fines < 5:
                print('this branch uses Cu/Cc and is not coded yet')
            elif fines >= 5 and fines <= 12:
                print('this branch uses Cu/Cc and is not coded yet')
            else: #fines > 12%
                #this basically says: if ML or MH (under A Line)
                if (liquid_limit < 25.5 and plasticity_index < 4) or (liquid_limit >= 25.5 and plasticity_index < (0.73*(liquid_limit - 20))):
                    sucs_group = 'Group symbol: SM'
                    if sand < 15:
                        sucs_class = 'Silty sand'
                    else: #sand>=15%
                        sucs_class = 'Silty sand with gravel'
                #this basically says: if CL or CH (above A line)
                elif (plasticity_index > 7 and plasticity_index >= (0.73*(liquid_limit - 20))):
                    sucs_group = 'Group symbol: SC'
                    if sand < 15:
                        sucs_class = 'Clayey sand'
                    else: #sand>=15%
                        sucs_class = 'Clayey sand with sand'
                #this basically says: if CL-ML (grey box)
                elif (4 <= plasticity_index <= 7 and plasticity_index >= (0.73*(liquid_limit - 20))):
                    sucs_group = 'Group symbol: SC-SM'
                    if sand < 15:
                        sucs_class = 'Silty, clayey sand'
                    else: #sand>=15%
                        sucs_class = 'Silty, clayey sand with gravel'
                else:
                    print('breaks here HHHHH')
    

    return sucs_group, sucs_class, plasticity_index, retained_sieve_200
    

if __name__ == '__main__':

    liquid_limit = float(input('Enter the liquid limit of the sample'))
    plastic_limit = float(input('Enter the plastic limit of the sample'))
    gravel = float(input('Enter the porcent of gravel obtained'))
    sand = float(input('Enter the of sand obtained'))
    
    print('\nIs the soil organic?')
    while True:
        try:
            is_organic = input ('Input "true" or "false"').lower()
            if is_organic == 'true':
                print('The soil has organic contents')
                break       
            elif is_organic == 'false':
                print('The soil does not have organic contents')
                break
            else:
                print('ERROR: You must choose between true or false')
        except ValueError:
            print ('ERROR: You must choose between true or false')
            
    classification = sucs_grainsize(liquid_limit, plastic_limit, gravel, sand, is_organic)
    
    print()
    print('The SUCS group is: ' + str(classification[0]))
    print('The SUCS class is: ' + str(classification[1]))
#     print('The PI is: ' + str(classification[2]))
#     print('The retained in sieve 200 is: ' + str(classification[3]))

```
